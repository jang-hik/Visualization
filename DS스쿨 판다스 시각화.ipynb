```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

* 판다스의 시리즈나 데이터프레임음 plot이라는 시각화 메서드를 내장하고 있다.
* plot은 matplotlib을 내부에서 임포트하여 사용한다.
* plot 메서드의 kind라는 인수를 바꾸면 여러가지 plot을 그릴 수 있다.
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot (DataFrame only)
    - 'hexbin' : hexbin plot (DataFrame only)

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1
```

```python
df1.plot(figsize=(15,5))
plt.xlabel('time')
plt.ylabel('Data')
plt.show()
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
```

```python
iris.sepal_length[:50].plot(kind='bar', rot=0, figsize=(15,5))
plt.xlabel('row')
plt.ylabel('sepal_length')
plt.show()
```

```python
# kind 인수에 문자열을 쓰는 대신 plot.bar처럼 직접 메서드로 사용할 수도 있다.
iris[:10].plot.bar(rot=0, figsize=(15,5))
plt.xlabel('row')
plt.ylabel('Feature value')
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0, figsize=(15,6))
plt.ylabel('mean')
```

```python
df2.T
```

```python
df2.T.plot.bar(rot=0, figsize=(15,6))
plt.ylabel('mean')
```

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('Percentage of passengers per class')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
iris.plot.hist(ax=ax1)
iris.plot.kde(ax=ax2)
plt.show()
```

```python
iris.plot.box()
plt.show()
```

```python
# 박스플롯에 대해서는 추가적인 기능을 가진 boxplot 명령이 별도로 있다.
iris.boxplot(by='species', figsize=(15,8))
plt.show()
```
